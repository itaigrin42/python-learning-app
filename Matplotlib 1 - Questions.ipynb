1. Draw a line with suitable label in the x axis, y axis and a title

```python
x = [0,10,20,30,40,50]
y = [0,20,40,60,80,100]
plt.title("Line")
plt.plot(x,y)
plt.show()
```

2. Draw a line using given axis values with suitable label in the x axis , y axis and a title.

```python
x = [0,10,20,30,40,50]
y = [0,20,40,30,20,10]
plt.title("Graph!")
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
plt.plot(x,y)
plt.show()
```

3. Plot 2 lines on same plot with suitable legends of each line <br>
#### hint ( plt.legend() )

```python
x1 = [0,3,6,8]
y1 = [0,50,200,100]

x2 = [1,4,5,7]
y2 = [0,40,150,75]
plt.title("2 lines with suitable legends")
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
plt.plot(x1,y1,label = 'line 1')
plt.plot(x2,y2,label = 'line 2')
plt.legend()
plt.show()
```

4. Plot 2 lines with legends, different widths and colors

```python
plt.title("2 lines with suitable legends")
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
plt.plot(x1,y1, c = '#5CAF50', lw = '5',label = 'line 1')
plt.plot(x2,y2, c = '#8F00FF', lw = '3',label = 'line 2')
plt.legend()
plt.show()
```

5. Plot 2 lines with different styles

```python
plt.title("2 lines with suitable legends")
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
plt.plot(x1,y1, c = '#5CAF50', lw = '5', ls = 'dashed', label = 'line 1')
plt.plot(x2,y2, c = '#8F00FF', lw = '3', ls = ':', label = 'line 2')
plt.legend()
plt.show()
```

6. Draw a line and set the line markers

```python
x = [0,10,20,30,40,50]
y = [0,20,40,30,20,10]
plt.title("Marker")
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
plt.plot(x,y,marker = 'o',ms = 10, mec = 'b', mfc = 'y',ls = '-.',lw = '3')
plt.show()
```

7. Display the current axis limits values and set new axis values
