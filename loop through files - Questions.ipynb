```python
SourcePath = 'C:\\Users\\itaig\\Desktop\\Data'
ProcessPath = 'C:\\Users\\itaig\\Desktop\\Data\\Process'
DonePath = 'C:\\Users\\itaig\\Desktop\\Data\\Done'
```

```python
# Getting the files names from the specified directoy
# ("r" means the string will be treated as raw string)
DataSource = os.listdir(SourcePath)
DataSource
```

```python
import pandas as pd
```

```python
# Write the code
now = datetime.now()
safe_now = str(now).replace(":", "-")
df_list = []
for file_name in DataSource:
    if file_name.find(".csv") != -1:
        shutil.move(f'{SourcePath}\\{file_name}',f'{ProcessPath}\\{file_name}')
        df = pd.read_csv(f'{ProcessPath}\\{file_name}')
        df_list.append(df)
        final_df = pd.concat(df_list, ignore_index=True)
        shutil.move(f'{ProcessPath}\\{file_name}',f'{DonePath}\\{file_name}')
        os.rename(f'{DonePath}\\{file_name}',f'{DonePath}\\{safe_now}_{file_name}')
```
