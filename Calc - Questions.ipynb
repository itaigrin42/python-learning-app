```python
def add(x, y):
    return x+y
```

```python
def Subtract(x, y):
    return x-y
```

```python
def Multiply(x, y):
    return x*y
```

```python
def Divide(x, y):
    return x/y
```

```python
def calc():
    
    print('Select operation:')
    print('1.Add')
    print('2.Subtract')
    print('3.Multiply')
    print('4.Divide')

    a = int(input('Enter choice(1/2/3/4):'))
    b = int(input('Enter first number'))
    c = int(input('Enter second number'))
    if a==1:
        print(b,"+",c,"=", add(b,c))
    elif a==2:
        print(b,"-",c,"=", Subtract(b,c))
    elif a==3:
        print(b,"*",c,"=", Multiply(b,c))
    elif a==4:
        print(b,"/",c,"=", Divide(b,c))
    else:
        print("choose again")
```
