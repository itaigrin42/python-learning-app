**Exercise 1**

Create a DataFrame from a dictionary.


```python
import pandas as pd
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
print(df)
```

**Exercise 2**

Display the first 5 rows of a DataFrame.


```python
import pandas as pd
df = pd.DataFrame({'A': range(10), 'B': range(10, 20)})
print(df.head())
```

**Exercise 3**

Select a single column.


```python
import pandas as pd
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
print(df['A'])
```

**Exercise 4**

Filter rows based on a condition.


```python
import pandas as pd
df = pd.DataFrame({'A': [1, 2, 3, 4], 'B': [10, 20, 30, 40]})
print(df[df['A'] > 2])
```

**Exercise 5**

Add a new calculated column.


```python
import pandas as pd
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df['C'] = df['A'] + df['B']
print(df)
```

**Exercise 6**

Drop a column.


```python
import pandas as pd
df = pd.DataFrame({'A': [1, 2], 'B': [3, 4], 'C': [5, 6]})
print(df.drop('B', axis=1))
```

**Exercise 7**

Rename columns.


```python
import pandas as pd
df = pd.DataFrame({'a': [1], 'b': [2]})
print(df.rename(columns={'a': 'A', 'b': 'B'}))
```

**Exercise 8**

Check for missing values.


```python
import pandas as pd
df = pd.DataFrame({'A': [1, None, 3], 'B': [4, 5, None]})
print(df.isna().sum())
```

**Exercise 9**

Fill missing values.


```python
import pandas as pd
df = pd.DataFrame({'A': [1, None, 3]})
print(df.fillna(0))
```

**Exercise 10**

Sort by a column.


```python
import pandas as pd
df = pd.DataFrame({'A': [3, 1, 2], 'B': [6, 4, 5]})
print(df.sort_values('A'))
```

**Exercise 11**

Reset the index.


```python
import pandas as pd
df = pd.DataFrame({'A': [1, 2, 3]}, index=[10, 20, 30])
print(df.reset_index())
```

**Exercise 12**

Set a column as index.


```python
import pandas as pd
df = pd.DataFrame({'id': [1, 2, 3], 'val': [10, 20, 30]})
print(df.set_index('id'))
```

**Exercise 13**

Count unique values in a column.


```python
import pandas as pd
df = pd.DataFrame({'A': [1, 2, 2, 3, 3, 3]})
print(df['A'].nunique())
```

**Exercise 14**

Convert a column to datetime format.


```python
import pandas as pd
df = pd.DataFrame({'date': ['2024-01-01', '2024-01-02']})
df['date'] = pd.to_datetime(df['date'])
print(df.dtypes)
```

**Exercise 15**

Export a DataFrame to CSV.


```python
import pandas as pd
df = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df.to_csv('output.csv', index=False)
```
