```python
lst = [12, 45, 7, 89, 34]
print(max(lst), min(lst))
```

**Exercise 4**

Given the list [9, 3, 7, 1, 5], create a new sorted list in ascending order without modifying the original list.


```python
lst = [9, 3, 7, 1, 5]
new_lst = sorted(lst)
print(new_lst)
```

**Exercise 6**

**Exercise 7**

Print each fruit from ['apple', 'banana', 'cherry'] together with its position number.


```python
fruits = ['apple', 'banana', 'cherry']
for i, f in enumerate(fruits, 1):
    print(i, f)
```

**Exercise 8**

Combine the lists ['John', 'Emma', 'Liam'] and [25, 30, 22] into pairs of name and age.


```python
names = ['John', 'Emma', 'Liam']
ages = [25, 30, 22]
print(list(zip(names, ages)))
```

**Exercise 9**

From the list [2, 4, 6, 8], create a new list containing the square of each number.


```python
lst = [2, 4, 6, 8]
print([x**2 for x in lst])
```

**Exercise 10**

From the list [11, 18, 25, 42, 57, 60], create a new list containing only the even numbers.


```python
lst = [11, 18, 25, 42, 57, 60]
print([x for x in lst if x % 2 == 0])
```

**Exercise 11**

**Exercise 12**

Check whether all numbers in [5, 12, 8, 3] are greater than zero.


```python
lst = [5, 12, 8, 3]
print(all(x > 0 for x in lst))
```
