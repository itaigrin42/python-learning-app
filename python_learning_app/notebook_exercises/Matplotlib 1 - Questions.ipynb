## Questions


**Exercise 1**

Create a basic line plot.


```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3], [4, 5, 6])
plt.show()
```

**Exercise 2**

Add title and axis labels to a plot.


```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3], [4, 5, 6])
plt.title('My Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

**Exercise 3**

Plot two lines on the same graph.


```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3], [4, 5, 6], label='Line 1')
plt.plot([1, 2, 3], [6, 5, 4], label='Line 2')
plt.legend()
plt.show()
```

**Exercise 4**

Create a scatter plot.


```python
import matplotlib.pyplot as plt
plt.scatter([1, 2, 3], [4, 5, 6])
plt.show()
```

**Exercise 5**

Customize marker styles and sizes.


```python
import matplotlib.pyplot as plt
plt.scatter([1, 2, 3], [4, 5, 6], marker='s', s=100)
plt.show()
```

**Exercise 6**

Create a bar chart.


```python
import matplotlib.pyplot as plt
plt.bar(['A', 'B', 'C'], [3, 7, 5])
plt.show()
```

**Exercise 7**

Create a histogram with 20 bins.


```python
import matplotlib.pyplot as plt
import numpy as np
data = np.random.randn(1000)
plt.hist(data, bins=20)
plt.show()
```

**Exercise 8**

Change line color and style.


```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3], [4, 5, 6], color='red', linestyle='--')
plt.show()
```

**Exercise 9**

Add a legend to a plot.


```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3], [4, 5, 6], label='Data')
plt.legend()
plt.show()
```

**Exercise 10**

Adjust figure size.


```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.plot([1, 2, 3], [4, 5, 6])
plt.show()
```

**Exercise 11**

Set axis limits manually.


```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3], [4, 5, 6])
plt.xlim(0, 5)
plt.ylim(0, 10)
plt.show()
```

**Exercise 12**

Add grid lines.


```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3], [4, 5, 6])
plt.grid(True)
plt.show()
```

**Exercise 13**

Save a plot as a PNG file.


```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3], [4, 5, 6])
plt.savefig('plot.png')
plt.show()
```

**Exercise 14**

Rotate x-axis labels.


```python
import matplotlib.pyplot as plt
plt.bar(['Jan', 'Feb', 'Mar'], [1, 2, 3])
plt.xticks(rotation=45)
plt.show()
```

**Exercise 15**

Create subplots with 2 rows and 1 column.


```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(2, 1)
axes[0].plot([1, 2, 3], [4, 5, 6])
axes[1].plot([1, 2, 3], [6, 5, 4])
plt.tight_layout()
plt.show()
```
