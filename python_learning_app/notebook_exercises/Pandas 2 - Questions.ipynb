**Exercise 1**

Group data by a categorical column.


```python
import pandas as pd
df = pd.DataFrame({'cat': ['A', 'B', 'A', 'B'], 'val': [1, 2, 3, 4]})
print(df.groupby('cat').sum())
```

**Exercise 2**

Calculate mean and sum after grouping.


```python
import pandas as pd
df = pd.DataFrame({'cat': ['A', 'B', 'A'], 'val': [1, 2, 3]})
print(df.groupby('cat').agg(['mean', 'sum']))
```

**Exercise 3**

Merge two DataFrames.


```python
import pandas as pd
df1 = pd.DataFrame({'id': [1, 2], 'a': [10, 20]})
df2 = pd.DataFrame({'id': [1, 2], 'b': [30, 40]})
print(pd.merge(df1, df2, on='id'))
```

**Exercise 4**

Concatenate multiple DataFrames.


```python
import pandas as pd
df1 = pd.DataFrame({'A': [1, 2]})
df2 = pd.DataFrame({'A': [3, 4]})
print(pd.concat([df1, df2], ignore_index=True))
```

**Exercise 5**

Apply a custom function to a column.


```python
import pandas as pd
df = pd.DataFrame({'A': [1, 2, 3]})
df['B'] = df['A'].apply(lambda x: x * 2)
print(df)
```

**Exercise 6**

Create a pivot table.


```python
import pandas as pd
df = pd.DataFrame({'cat': ['A', 'B', 'A'], 'sub': ['X', 'X', 'Y'], 'val': [1, 2, 3]})
print(pd.pivot_table(df, values='val', index='cat', columns='sub'))
```

**Exercise 7**

Replace specific values in a column.


```python
import pandas as pd
df = pd.DataFrame({'A': [1, 2, 3, 2]})
print(df['A'].replace(2, 99))
```

**Exercise 8**

Drop duplicate rows.


```python
import pandas as pd
df = pd.DataFrame({'A': [1, 2, 2, 3]})
print(df.drop_duplicates())
```

**Exercise 9**

Select data using label-based indexing.


```python
import pandas as pd
df = pd.DataFrame({'A': [1, 2, 3]}, index=['x', 'y', 'z'])
print(df.loc['x'])
```

**Exercise 10**

Select data using position-based indexing.


```python
import pandas as pd
df = pd.DataFrame({'A': [1, 2, 3, 4, 5]})
print(df.iloc[1:4])
```

**Exercise 11**

Perform rolling average calculation.


```python
import pandas as pd
df = pd.DataFrame({'A': [1, 2, 3, 4, 5]})
print(df['A'].rolling(3).mean())
```

**Exercise 12**

Calculate correlation between columns.


```python
import pandas as pd
df = pd.DataFrame({'A': [1, 2, 3], 'B': [2, 4, 6]})
print(df.corr())
```

**Exercise 13**

Count frequency of values in a column.


```python
import pandas as pd
df = pd.DataFrame({'A': [1, 2, 2, 3, 2]})
print(df['A'].value_counts())
```

**Exercise 14**

Sort by multiple columns.


```python
import pandas as pd
df = pd.DataFrame({'A': [1, 1, 2], 'B': [3, 2, 1]})
print(df.sort_values(['A', 'B']))
```

**Exercise 15**

Create a frequency distribution table.


```python
import pandas as pd
df = pd.DataFrame({'A': [1, 2, 2, 3, 3, 3]})
print(df['A'].value_counts().sort_index())
```
