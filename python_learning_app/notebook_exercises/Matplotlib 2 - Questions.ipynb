## Questions


**Exercise 1**

Create a pie chart.


```python
import matplotlib.pyplot as plt
plt.pie([30, 50, 20], labels=['A', 'B', 'C'])
plt.show()
```

**Exercise 2**

Highlight one slice of a pie chart.


```python
import matplotlib.pyplot as plt
explode = (0, 0.1, 0)
plt.pie([30, 50, 20], labels=['A', 'B', 'C'], explode=explode)
plt.show()
```

**Exercise 3**

Create a horizontal bar chart.


```python
import matplotlib.pyplot as plt
plt.barh(['A', 'B', 'C'], [3, 7, 5])
plt.show()
```

**Exercise 4**

Customize tick labels.


```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3], [4, 5, 6])
plt.xticks([1, 2, 3], ['One', 'Two', 'Three'])
plt.show()
```

**Exercise 5**

Plot data from a CSV file.


```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('data.csv')
plt.plot(df['x'], df['y'])
plt.show()
```

**Exercise 6**

Use different line styles in one plot.


```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3], [4, 5, 6], linestyle='-')
plt.plot([1, 2, 3], [5, 6, 7], linestyle='--')
plt.show()
```

**Exercise 7**

Add text annotations to a plot.


```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3], [4, 5, 6])
plt.annotate('peak', xy=(2, 5), xytext=(2.5, 6), arrowprops=dict(arrowstyle='->'))
plt.show()
```

**Exercise 8**

Create multiple subplots with shared axes.


```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(2, 1, sharex=True)
axes[0].plot([1, 2, 3], [4, 5, 6])
axes[1].plot([1, 2, 3], [6, 5, 4])
plt.show()
```

**Exercise 9**

Plot cumulative sum data.


```python
import matplotlib.pyplot as plt
import numpy as np
data = [1, 2, 3, 4, 5]
plt.plot(np.cumsum(data))
plt.show()
```

**Exercise 10**

Create a stacked bar chart.


```python
import matplotlib.pyplot as plt
import numpy as np
x = ['A', 'B', 'C']
y1 = [3, 5, 2]
y2 = [2, 4, 3]
plt.bar(x, y1)
plt.bar(x, y2, bottom=y1)
plt.show()
```

**Exercise 11**

Plot error bars.


```python
import matplotlib.pyplot as plt
plt.errorbar([1, 2, 3], [4, 5, 6], yerr=[0.5, 0.5, 0.5])
plt.show()
```

**Exercise 12**

Use logarithmic scale on axes.


```python
import matplotlib.pyplot as plt
plt.semilogy([1, 2, 3], [10, 100, 1000])
plt.show()
```

**Exercise 13**

Customize font sizes.


```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3], [4, 5, 6])
plt.title('Title', fontsize=16)
plt.xlabel('X', fontsize=12)
plt.show()
```

**Exercise 14**

Remove top and right borders of a plot.


```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3], [4, 5, 6])
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.show()
```

**Exercise 15**

Create a heatmap.


```python
import matplotlib.pyplot as plt
import numpy as np
data = np.random.rand(5, 5)
plt.imshow(data, cmap='viridis')
plt.colorbar()
plt.show()
```
