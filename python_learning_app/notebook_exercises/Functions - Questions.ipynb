**Exercise 1**

Write a function that returns the square of a number.


```python
def square(n):
    return n ** 2
print(square(5))
```

**Exercise 2**

Write a function that checks if a number is prime.


```python
def is_prime(n):
    return n > 1 and all(n % i for i in range(2, int(n**0.5)+1))
print(is_prime(17))
```

**Exercise 3**

Create a function that returns the average of a list.


```python
def avg(lst):
    return sum(lst) / len(lst)
print(avg([1, 2, 3, 4, 5]))
```

**Exercise 4**

Write a function with default parameter values.


```python
def greet(name='World'):
    return f'Hello, {name}'
print(greet('Alice'))
```

**Exercise 5**

Create a function that returns multiple values.


```python
def min_max(lst):
    return min(lst), max(lst)
a, b = min_max([1, 5, 3])
print(a, b)
```

**Exercise 6**

Write a recursive function to calculate factorial.


```python
def factorial(n):
    return 1 if n <= 1 else n * factorial(n-1)
print(factorial(5))
```

**Exercise 7**

Create an anonymous function that multiplies two numbers.


```python
mul = lambda x, y: x * y
print(mul(3, 4))
```

**Exercise 8**

Write a function that counts vowels in a string.


```python
def count_vowels(s):
    return sum(1 for c in s.lower() if c in 'aeiou')
print(count_vowels('hello'))
```

**Exercise 9**

Create a function that removes duplicates from a list.


```python
def remove_dupes(lst):
    return list(dict.fromkeys(lst))
print(remove_dupes([1, 2, 2, 3]))
```

**Exercise 10**

Write a function that accepts unlimited positional arguments.


```python
def sum_all(*args):
    return sum(args)
print(sum_all(1, 2, 3, 4))
```

**Exercise 11**

Write a function that accepts unlimited keyword arguments.


```python
def print_kwargs(**kwargs):
    for k, v in kwargs.items():
        print(k, v)
print_kwargs(a=1, b=2)
```

**Exercise 12**

Create a function that sorts a list of dictionaries by a key.


```python
def sort_by_key(lst, key):
    return sorted(lst, key=lambda x: x[key])
data = [{'n': 2}, {'n': 1}, {'n': 3}]
print(sort_by_key(data, 'n'))
```

**Exercise 13**

Write a function that validates email format.


```python
import re
def valid_email(s):
    return bool(re.match(r'[^@]+@[^@]+\.[^@]+', s))
print(valid_email('a@b.com'))
```

**Exercise 14**

Create a decorator that measures execution time.


```python
import time
def timer(f):
    def wrap(*a,**k):
        t=time.perf_counter(); r=f(*a,**k); print(time.perf_counter()-t); return r
    return wrap
@timer
def f(): time.sleep(0.1)
f()
```

**Exercise 15**

Write a function that reads a file and returns the number of lines.


```python
def count_lines(path):
    with open(path) as f: return len(f.readlines())
print(count_lines('file.txt'))
```
