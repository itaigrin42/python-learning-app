**Exercise 1**

Check if a number is positive, negative, or zero.


```python
n = int(input('Enter number: '))
if n > 0: print('positive')
elif n < 0: print('negative')
else: print('zero')
```

**Exercise 2**

Determine if a year is a leap year.


```python
year = int(input('Year: '))
is_leap = (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)
print(is_leap)
```

**Exercise 3**

Print numbers from 1 to 20.


```python
for i in range(1, 21):
    print(i)
```

**Exercise 4**

Print only odd numbers between 1 and 50.


```python
for i in range(1, 51, 2):
    print(i)
```

**Exercise 5**

Calculate the sum of numbers until the user enters 0.


```python
total = 0
while True:
    n = int(input('Enter number (0 to stop): '))
    if n == 0: break
    total += n
print(total)
```

**Exercise 6**

Stop a loop when a number greater than 100 is found.


```python
lst = [50, 75, 120, 30]
for x in lst:
    if x > 100: break
    print(x)
```

**Exercise 7**

Skip negative numbers in a list.


```python
lst = [1, -2, 3, -4, 5]
for x in lst:
    if x < 0: continue
    print(x)
```

**Exercise 8**

Create a simple number guessing game.


```python
import random
secret = random.randint(1, 10)
while int(input('Guess: ')) != secret:
    print('Try again')
print('Correct!')
```

**Exercise 9**

Print the multiplication table of a given number.


```python
n = int(input('Number: '))
for i in range(1, 11):
    print(f'{n} x {i} = {n*i}')
```

**Exercise 10**

Create a dictionary with numbers and their squares.

**Exercise 13**

Print all keys and values from a dictionary.


```python
d = {'a': 1, 'b': 2}
for k, v in d.items():
    print(k, v)
```

**Exercise 14**

Count vowels in a string.


```python
s = 'hello world'
vowels = sum(1 for c in s if c in 'aeiou')
print(vowels)
```

**Exercise 15**

Find the first prime number between 50 and 100.


```python
def is_prime(n):
    return all(n % i for i in range(2, int(n**0.5)+1)) and n > 1
for n in range(50, 101):
    if is_prime(n):
        print(n); break
```
