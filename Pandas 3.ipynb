## Using `lambda` Functions with `df`

<br>
<br>

```python
lambda x : x * 0.9
```

```python
(lambda x : x * 0.9)(100)
```

```python
df
```

```python
# Anonymus
df['RevAfterDiscount'] = df.Rev.apply(lambda x : x * 0.9)
df
```

<br>

<font size="5">We can also do that with `def` function</font>

```python
def discount(x):
    return x * 0.9
```

```python
df['RevAfterDiscount_2'] = df.Rev.apply(discount)
df
```

<br>

<font size="6">Another Example</font>

```python
df = df[['OrderID','ProductID','Qty','Rev']]
df
```

<br>

<font size="4">Creating another column **OnSale**</font>
<br>
<font size="4" color='red'>The logic doesn't matter</font>

```python
df.loc[:,'OnSale'] = df.index % 2 != 0
df
```

<br>

<font size="5">Simple example of how to write a <u>**One Liner `IF`**</u></font>

<br>

<font size="4">NOT - <u>**One Liner `IF`**</u></font>

```python
row = df.loc[1]

if row['OnSale']:
     print(row['Rev'] * 0.5)
else:
    print(row['Rev'])
```

<br>

<font size="4">YES - <u>**One Liner `IF`**</u></font>

```python
row = df.loc[1]

print(row['Rev'] * 0.5) if row['OnSale'] else print(row['Rev'])
```

<br>

<font size="5">Using the <u>**One Liner `IF`**</u> with `lambda` function</font>

```python
(lambda row :   row['Rev'] * 0.5 if row['OnSale'] else row['Rev'])
```

```python
row = df.loc[1]

(lambda row :   row['Rev'] * 0.5 if row['OnSale'] else row['Rev'])(row)
```

<br>

<font size="5">Using the <u>**One Liner `IF`**</u> with `lambda` function - <u>WITH A `df`</u></font>

```python
discount = 0.5

df['discounted_rev'] = df.apply(  lambda row   :   row['Rev'] * discount if row['OnSale'] else row['Rev'], axis=1  )
df
```

```python
df.drop('discounted_rev',axis=1, inplace=True)
df
```

## Using `lambda` Functions with `groupby()`

<br>
<br>

```python
grouped = df.groupby('OrderID', as_index=False)
grouped
```

```python
grouped.Rev.sum()
```

```python
df
```

```python
transformed = grouped.Rev.transform(lambda x : x.sum())
transformed
```

```python
# Changing Series 'name', for the 'join', so we dont have 2 columns with the same name
transformed.name = 'OrderRev'
transformed.name
```

```python
df2 = df.join(transformed)
df2
```

```python
df2['SalePer%FromOrder'] = df2['Rev'] / df2['OrderRev']
df2[['OrderID', 'ProductID', 'Qty', 'Rev', 'OrderRev', 'SalePer%FromOrder']]
```
