### Exercise 3

Group by `Fruit` column and create `DataFrameGroupBy`

````python
grouped=df.groupby('Fruit')
grouped
````

### Exercise 5

Count each row per `Fruit` column (Keep `Fruit` column as a REGULAR column) 

````python
df.groupby('Fruit', as_index=False).Fruit.size()
````

### Exercise 6

### Exercise 7

Group by `Fruit` and `FruitColor` columns, calculate 2 colums (`Sold` & `InStock`) with `sum` function (Keep `Fruit` and `FruitColor` columns as a REGULAR columns)

````python
df.groupby(['Fruit','FruitColor'], as_index=False)[['Sold','InStock']].sum()
````

### Exercise 8

Group by Fruit and FruitColor columns, calculate 2 colums (Sold & InStock) with "sum" and "mean" functions

Group by `Fruit` and `FruitColor` columns, calculate 2 colums:

- `Sold` with `sum` and `mean` functions
- `InStock` with `min` and `max` functions

(Keep `Fruit` and `FruitColor` columns as a REGULAR columns)

````python
g = df.groupby(['Fruit','FruitColor'], as_index=False).agg({'Sold':['sum','mean'],'InStock':['min','max']})
g
````
