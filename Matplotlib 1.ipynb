Matplotlib is a graphical library that produce publication quality plots. Those plots can be saved in many known formats like gif, png, jpg, pdf, eps, postcript, etc.
<br>This library depends on numpy.
<br>Here is an example of how to create a plot with matplotlib:

Let's start by importing `Matplotlib.pyplot` as follows:

```python
# we are using the inline backend
#%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
x = [0,3,6,8]
y = [0,50,200,100]


print(x)
print(y)
```

```python
# Create a line in a diagram from x,y positions =  (0,0) to (8,100):
# "plot()" is the function to create the diagram
# The function takes parameters for specifying points in the diagram.
# The FIRST Parameter is an array containing the points on the x-axis.
# The FIRST Parameter is an array containing the points on the y-axis.
plt.plot(x,y)
```

```python
# if will add "plt.show()" it as if we print the diagram
plt.plot(x,y)
plt.show()
```

```python
# "plt.show()" in the code cell alone --> will do nothing
plt.show()
```

## Marker
- 'o'	Circle	
- '*'	Star	
- '.'	Point	
- ','	Pixel	
- 'x'	X	
- 'X'	X (filled)	
- '+'	Plus	
- 'P'	Plus (filled)	
- 's'	Square	
- 'D'	Diamond	
- 'd'	Diamond (thin)	
- 'p'	Pentagon	
- 'H'	Hexagon	
- 'h'	Hexagon	
- 'v'	Triangle Down	
- '^'	Triangle Up	
- '<'	Triangle Left	
- '>'	Triangle Right	
- '1'	Tri Down	
- '2'	Tri Up	
- '3'	Tri Left	
- '4'	Tri Right	
- '|'	Vline	
- '_'	Hline

```python
# We can choose to plot with something different than a line
plt.plot(x, y, 'o')
plt.show()
```

```python
# We can choose to plot with something different than a line
plt.plot(x, y, 'D')
plt.show()
```

```python
# We can choose to plot with something different than a line
plt.plot(x, y, '+')
plt.show()
```

```python
# If we'll use the keyword "marker" it will draw a line and emphasize each point
plt.plot(x, y, marker = 'D')
plt.show()
```

```python
# marker size = ms
# marker edge color = mec
# marker face color = mfc

plt.plot(x, y,  marker = '*', ms = 20, mec = 'y', mfc = 'r')
plt.show()
```

```python
# for more colors, just google "color code"
# or use this link:
# https://htmlcolorcodes.com/

plt.plot(x, y,  marker = '*', ms = 20, mec = '#8F00FF', mfc = '#CE14DE')
plt.show()
```

## Line
- 'solid' (default)	'-'	
- 'dotted'	':'	
- 'dashed'	'--'	
- 'dashdot'	'-.'	
- 'None'	'' or ' '

```python
# linestyle

plt.plot(x, y, ls = 'dotted', marker = '*', ms = 20, mec = '#8F00FF', mfc = '#FFC300')
plt.show()
```

```python
# linestyle

plt.plot(x, y, ls = 'dashed', marker = '*', ms = 20, mec = '#8F00FF', mfc = '#FFC300')
plt.show()
```

```python
# linestyle
# Instead od "dotted" use ":"

plt.plot(x, y, ls = ':', marker = '*', ms = 20, mec = '#8F00FF', mfc = '#FFC300')
plt.show()
```

```python
# line color

plt.plot(x, y, ls = ':', c = 'r', marker = '*', ms = 20,  mec = '#8F00FF', mfc = '#FFC300')
plt.show()
```

```python
# line color
# Hexadecimal color values
plt.plot(x, y, ls = ':', c = '#5CAF50', marker = '*', ms = 20,  mec = '#8F00FF', mfc = '#FFC300')
plt.show()
```

```python
# line Width
plt.plot(x, y, ls = ':', c = '#5CAF50', lw = '5', marker = '*', ms = 20,  mec = '#8F00FF', mfc = '#FFC300')
plt.show()
```

## fmt

```python
# fmt
# Stands for: shortcut string notation parameter to specify the marker
# it comtrols --> "marker"|"line"|"color"
plt.plot(x, y, '*:g')
plt.show()
# '*' = + for each point
# ':' = Dotted line
# 'g' = green
```

## Multiple Lines

```python
# Multiple Lines
# plt.plot() function for each line
x1 = [0,3,6,8]
y1 = [0,50,200,100]

x2 = [1,4,5,7]
y2 = [0,40,150,75]

plt.plot(x1, y1)
plt.plot(x2, y2)

plt.show()
```

```python
# Multiple Lines
# Another way is to specifiyng the x and y point values for both lines
plt.plot(x1, y1,x2,y2)
plt.show()
```

## Title & Lables

```python
# Lables
plt.plot(x1, y1,x2,y2)

plt.title("Netflix")
plt.xlabel("Year")
plt.ylabel("Users in M")


plt.show()

```

```python
# Lables
# Add properties

plt.plot(x1, y1,x2,y2)

f1 = {'family':'serif','color':'red','size':30}
f2 = {'family':'serif','color':'darkblue','size':15}

plt.title("Netflix", fontdict = f1)
plt.xlabel("Year", fontdict = f2)
plt.ylabel("Users in M", fontdict = f2)


plt.show()
```

```python
# Lables - Positioin


plt.plot(x1, y1,x2,y2)

f1 = {'family':'serif','color':'red','size':30}
f2 = {'family':'serif','color':'darkblue','size':15}

plt.title("Netflix", fontdict = f1, loc = 'left')
plt.xlabel("Year", fontdict = f2, loc = 'center')
plt.ylabel("Users in M", fontdict = f2, loc = 'top')


plt.show()
```

## Grid

```python
# Grid

plt.plot(x1, y1,x2,y2)

f1 = {'family':'serif','color':'red','size':30}
f2 = {'family':'serif','color':'darkblue','size':15}

plt.title("Netflix", fontdict = f1, loc = 'left')
plt.xlabel("Year", fontdict = f2, loc = 'center')
plt.ylabel("Users in M", fontdict = f2, loc = 'top')

plt.grid()

plt.show()
```

```python
# Grid

plt.plot(x1, y1,x2,y2)

f1 = {'family':'serif','color':'red','size':30}
f2 = {'family':'serif','color':'darkblue','size':15}

plt.title("Netflix", fontdict = f1, loc = 'left')
plt.xlabel("Year", fontdict = f2, loc = 'center')
plt.ylabel("Users in M", fontdict = f2, loc = 'top')

plt.grid(axis = 'x')

plt.show()
```

```python
# Grid

plt.plot(x1, y1,x2,y2)

f1 = {'family':'serif','color':'red','size':30}
f2 = {'family':'serif','color':'darkblue','size':15}

plt.title("Netflix", fontdict = f1, loc = 'left')
plt.xlabel("Year", fontdict = f2, loc = 'center')
plt.ylabel("Users in M", fontdict = f2, loc = 'top')

plt.grid(axis = 'y')

plt.show()
```

```python
# Grid with properties

plt.plot(x1, y1, label = 'model 1')
plt.plot(x2,y2, label = 'model 2')

f1 = {'family':'serif','color':'red','size':30}
f2 = {'family':'serif','color':'darkblue','size':15}

plt.title("Netflix", fontdict = f1, loc = 'left')
plt.xlabel("Year", fontdict = f2, loc = 'center')
plt.ylabel("Users in M", fontdict = f2, loc = 'top')

plt.grid(color = 'green', linestyle = '--', linewidth = 1)


plt.legend()

plt.show()
```
