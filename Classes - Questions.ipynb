### Question 1

1. Write a class for cars. Each car has the following attributes:  
    + Company (Toyota, Hyundai, Fiat, Seat)
    + Model (['Corola','Prius','Auris'],['i10','i30','ioniq'],['Tipo','Panda','Punto'],['Ibiza','Leon'])
    + BodyStyle (Sedan, Wagon, Hatchback)
    + Doors (2,4,5)
    + Cylinders (4-8)
    + Hybrid (True,False)
 
   <br>
2. you should be able to create an instance even without setting the attributes class
   <br>
3. Create a method to retrieve and to set each of the attributes
   <br>
4. The method for setting the attributes,  should set them only if they are one of the values mentioned above, else should return a warning
   <br>
5. For attributes (Company, Model) - setting the company and model should be implemented in the same method, a model can only be set for the company it belongs to
   <br>
6. The method for retrieving the attributes should return them in upper letters (only for string values)

```python
class Cars:
    VALID_MODELS = {
        "Toyota": ["Corola", "Prius", "Auris"],
        "Hyundai": ["i10", "i30", "ioniq"],
        "Fiat": ["Tipo", "Panda", "Punto"],
        "Seat": ["Ibiza", "Leon"]
    }

    VALID_BodyStyle = {
        "Sedan", "Wagon", "Hatchback"
    }

    VALID_Doors = {
        2,4,5
    }

    VALID_Cylinders = {
        4,5,6,7,8
    }
    
    VALID_Hybrid = {
        True,False
    }
    
    
    def __init__(self):
        self.Company = ""
        self.Model = ""

    def set_Company_and_Model(self, Company, Model):
        if Company not in self.VALID_MODELS:
            self.Company = ""
            self.Model = ""
            return f"{Company} is not a supported company"

        if Model not in self.VALID_MODELS[Company]:
            self.Company = ""
            self.Model = ""
            return f"{Model} is not one of the models mentioned above for company {Company}"

        self.Company = Company
        self.Model = Model

    def get_Company_and_Model(self):
        return self.Company, self.Model

    def setBodyStyle(self,BodyStyle):
        if BodyStyle not in self.VALID_BodyStyle:
            self.BodyStyle = ""
            return f"{BodyStyle} is not one of the bodyStyle list mentioned above"
        self.BodyStyle = BodyStyle

    def getBodyStyle(self):
        return(self.BodyStyle)

    def setDoors(self,Doors):
        if Doors not in self.VALID_Doors:
            self.Doors = ""
            return f"{Doors} is not one of the doors list mentioned above"
        self.Doors = Doors

    def getDoors(self):
        return(self.Doors)

    def setCylinders(self,Cylinders):
        if Cylinders not in self.VALID_Cylinders:
            self.Cylinders = ""
            return f"{Cylinders} is not one of the cylinders list mentioned above"
        self.Cylinders = Cylinders

    def getCylinders(self):
        return(self.Cylinders)

    def setHybrid(self,Hybrid):
        if Hybrid not in self.VALID_Hybrid:
            self.Hybrid = ""
            return f"{Hybrid} is not one of the hybrid list mentioned above"
        self.Hybrid = Hybrid

    def getHybrid(self):
        return(self.Hybrid)
```
