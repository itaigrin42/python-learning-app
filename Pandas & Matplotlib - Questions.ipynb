### Exercise 1

Let's start by importing `Pandas`, `Matplotlib.pyplot`

````python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
````

### Exercise 2

Create a df using this dictionary:

```python
d = {'ProductName':['chai','chang','Tofu'],
     '1996':[1605, 3017, 1385],
     '1997':[4887, 7038, 6234],
     '1998':[6295, 6299, 371]}
```

````python
d = {'ProductName':['chai','chang','Tofu'],
     '1996':[1605, 3017, 1385],
     '1997':[4887, 7038, 6234],
     '1998':[6295, 6299, 371]}
df = pd.DataFrame(d)
df
````

### Exercise 3

Plot chai rev over time<br>
We'll do that in steps

First,
set the 'ProductName' column as the index
<br>Save i to a new `df2`

````python
df2 = df.set_index('ProductName')
df2
````

### Exercise 4

Remove the name of the index

````python
df2 = df2.rename_axis(None)
df2
````

### Exercise 5

Extract the data series for 'Chai' product.
<br>Set it to a new Series called `chai`

````python
chai=df2.loc['chai']
chai
````

### Exercise 6

Plot a line plot

````python
chai.plot()
plt.locator_params(axis='x', nbins = len(chai)) 
plt.title("Rev for Chai")
plt.xlabel("Year")
plt.ylabel("Rev")
plt.show()
````

### Exercise 7

Check the chai index data types, what are they ?

````python
type(chai.index[0])
````

### Exercise 8

Change the index type to int

````python
chai.index = chai.index.astype(int)
````

### Exercise 9

We can clearly notice the revenue for `CHAI` spiked up in 1997.<br>
Let's annotate this spike in the plot.<br>
hint: try to use `plt.text()`

````python
chai.plot()
plt.locator_params(axis='x', nbins = len(chai)) 
plt.text(1997,6000, 'Rev Spiked Up in 97')
plt.title("Rev for Chai")
plt.xlabel("Year")
plt.ylabel("Rev")
plt.show()
````

With just a few lines of code, we were able to quickly identify and visualize the spike!

### Exercise 10

We can easily add more products to line plot to make meaningful comparisons Rev from different products.

Let's compare the Rev of products "Chang" and "Tofu" from 1996 to 1998.

Get the data set for "Chang" and "Tofu", and display dataframe.
<br> Save it to `df_ct`

````python
df_ct=df2.iloc[1:3]
df_ct
````

### Exercise 11

*pandas* plots the indices on the x-axis and the columns as individual lines on the y-axis.
<br>
<br>Since `df_ct` is a dataframe with the `ProductName` as the index and `years` as the columns, we must first swap the row and columns.
<br> hint: Use `transpose` method

````python
df_ct=df_ct.transpose()
df_ct
````

### Exercise 12

Plot graph. We will explicitly specify line plot by passing in `kind` parameter to `plot()`.

````python
df_ct.plot(kind = 'line',)
plt.locator_params(axis='x', nbins = len(df_ct)) 
plt.title("Chang and Tofu Rev")
plt.xlabel("Year")
plt.ylabel("Rev")
plt.show()
````

*pandas* will auomatically graph the two products on the same graph.

### Exercise 13

Create  the following bar chart

````python
df_ct.plot(kind = 'bar',)
plt.locator_params(axis='x', nbins = len(df_ct)) 
plt.title("Chang and Tofu Rev")
plt.xlabel("Year")
plt.ylabel("Rev")
plt.show()
````
