Add a ColorMap to the scatter

````python
import numpy as np

x = np.random.randint(90, size=(90))
y = np.random.randint(90, size=(90))
colors = np.random.randint(90, size=(90))
sizes = 10 * np.random.randint(90, size=(90))

plt.scatter(x,y,c=colors,s = sizes,cmap = 'nipy_spectral', alpha = 0.5)
plt.colorbar()
plt.show()
````
