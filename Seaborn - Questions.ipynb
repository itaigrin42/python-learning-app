## Seaborn

Let's start by importing `Pandas`, `NumPy`, `Seaborn` as follows:

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.set_theme()
```

### Load file - "DataFrame for Python.xlsx"

```python
import pandas as pd

df = pd.read_excel(r'C:\Users\itaig\Desktop\pyt\DataFrame for Python.xlsx')
df
```

1. Create the visualization below

```python
sns.relplot(
    data=df,
    x="ProductAmt", y="Rev", col="RevType",
    hue="RevType", style="RevType", size="Rev",
)
```

2. Create the visualization below

```python
sns.relplot(
    data=df,
    x="ProductAmt", y="Rev", row="RevType",
    hue="RevType", style="RevType", size="Rev",
)
```

3. Create the visualization below

```python
sns.relplot(data=df, x="ProductAmt", y="Rev", hue="RevType", col="CategoryName", size="Rev", col_wrap=2)
```

4. Create the visualization below

```python
sns.relplot(
    data=df,
    x="ProductAmt", y="Rev",
    hue="RevType", style="RevType")
```

5. Create the visualization below

```python
sns.displot(data=df, x="ProductAmt", col="RevType", kde=True)
```

6. Create the visualization below

```python
sns.catplot(data=df, kind="swarm", x="RevType", y="Rev", hue="RevType")
```

7. Create the visualization below

```python
sns.relplot(data=df, kind="line", x="Year", y="Rev", col="RevType")
```
