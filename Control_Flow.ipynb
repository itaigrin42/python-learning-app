## if - elif - else

Conditional statements lets us to select a different action depending if a condition is true or false.


````python
a = 11


if(a > 10):
    b = 12
elif(a < 6):
    b = 4
else:
    b = 2
    b = 4

b
````

````python
# Why b equal 4 ??
# Because "b=4" isn't part of the "if/else" statement --> indentation!!!
a = 11


if(a > 10):
    b = 12
elif(a < 6):
    b = 4
else:
    b = 2
b = 4

b
````

````python
# nested "if" statement

a = 13


if a < 8:
    print('a smaller then 8')
elif a < 15:
    if a > 12:
        print("a smaller then 15 nut bigger than 12")
else:
    print("a not smaller than 8 or 15")


````

````python
a = {'Chevrolet':150000,'Fiat':104300,'Mazda':114000,'Toyota':125000}
car = 'Ferrari'
if(car in a.keys()):
    print("The price of a %s is: %d NIS" % (car, a[car]))
else:
    print("%s is not in your list" % car)
````

````python
car = 'Fiat'
if(car in a.keys()):
    print("The price of a %s is: %d NIS" % (car, a[car]))
else:
    print("%s is not in your list" % car)
````

Some important points to remark:
- In Python, at the end of a condition we put a colon sign(':')
- After a colon we go to the next line and have to be sure there is an indentation (using four spaces)
- If we use an IDE or in Jupyter notebooks, the identation is made automatically
- In Python the <i>'else if'</i> is written as <i>'elif'</i>
- The elif keyword is pythons way of saying "if the previous conditions were not true, then try this condition"
- The else keyword catches anything which isn't caught by the preceding conditions
- You can also have an else without the elif

## while

The "while statement":


````python
i = 10

while(i > 0):
    print(i)
    i -= 1
````

````python
i = 10

while(i > 0):

    if i == 4:
        break

    print(i)

    i -= 1
# The loop Stopped sooner because "break"
````

````python
i = 10

while(i > 0):

    if i == 4:
        i -= 1
        continue

    print(i)

    i -= 1

# The loop "skipped" value "4" because "continue"
````

## List Comprehension

 shorter syntax to create a new list based on the values of an existing list

````python
Cars = ['Chevrolet','Fiat','Mazda','Toyota']
CoolCars = []
````

````python
CoolCars = [i for i in Cars if "o" in i]
# CoolCars = [expression for item in iterable if condition == True]
print(CoolCars)
````

````python
CoolCars = [i for i in Cars]
print(CoolCars)
````

````python
CoolCars = [i for i in Cars if i == "toyota"]
# Python strings are case sensitive -- Empyt List
print(CoolCars)
````

````python
CoolCars = [i.lower() for i in Cars]
# Set the values in the new list to lower case:
print(CoolCars)
````

### List Comprehension with "if" "else"

````python
CoolCars = [i if "o" in i else 'x' for i in Cars]
CoolCars
````

### List Comprehension with "elif"

We can’t use elif in list comprehension as we use in a normal code we have to convert it into fragments of if-else according to the condition requirement.

````python
CoolCars = [i if "o" in i else i if 'M' in i else 'No "o" and No "M"' for i in Cars]
CoolCars
````

## Dictionary Comprehension

 shorter syntax to create a new dict based on the values of an existing dict

````python
d = {'Python':1000,'Sql':500,'Pbi':200}
d
````

````python
Tot = sum(d.values())

d_percentage = {k: round(v / Tot,2) for k, v in d.items()}
d_percentage
````

### Dictionary Comprehension with "if"

````python
Tot = sum(d.values())

d_percentage_only_for_python = {k: round(v / Tot,2) for (k, v) in d.items() if k == 'Python'}
d_percentage_only_for_python
````

### Dictionary Comprehension with "if" "else"

```python
{ (some_key if condition else default_key):(something_if_true if condition else something_if_false) for key, value in dict_.items() }
```

````python
Tot = sum(d.values())

d_percentage_only_for_python = { k :(round(v / Tot,2) if k == 'Python' else v) for k,v in d.items() }
d_percentage_only_for_python
````

````python
Tot = sum(d.values())

d_percentage_only_for_python = { (k if k == 'Python' else k + ' - is not python key') : (round(v / Tot,2) if k == 'Python' else v) for k,v in d.items() }
d_percentage_only_for_python
````

## None

The "None" keyword is used to define a null value, or Blank value (no value at all).
<br>
None is not the same as 0, False, or an empty string. None is a data type of its own (NoneType) and only None can be None.

````python
x = None

if x == True:
    print("None = True")
elif x == False:
    print ("None = False")
elif x == None:
    print("None IS None !!!")
````
