# Which Num

Let's use `while` loops to create a guessing game.

## The Challenge:

Write a program that picks a random integer from 1 to 100, and has players guess the number. The rules are:

1. If a player's guess is less than 1 or greater than 100, say "OUT OF BOUNDS"

2. On a player's first turn, if their guess is
 * within 10 of the number, return "WARM!"
 * further than 10 away from the number, return "COLD!"
3. On all subsequent turns, if a guess is 
 * closer to the number than the previous guess return "WARMER!"
 * farther from the number than the previous guess, return "COLDER!"
4. When the player's guess equals the number, tell them they've guessed correctly *and* how many guesses it took!


## Instructions

1. First, pick a random integer from 1 to 100 using the random module and assign it to a variable
<br>Note: `random.randint(a,b)` returns a random integer in range `[a, b]`, including both end points.

2. Write a `while` loop that compares the player's guess to our number. If the player guesses correctly, break from the loop.
<br>Otherwise, tell the player if they're warmer or colder, and continue asking for guesses.

<br>
<u>Some hints:</u><br>

* it may help to sketch out all possible combinations on paper first!
* you can use the `abs()` function to find the positive difference between two numbers
* if you append all new guesses to the list, then the previous guess is given as `guesses[-2]`

```python
import random

print("WELCOME TO 'Which Num'!")
print('')
print("I'm thinking of a number between 1 and 100")
print("If your FIRST guess is more than 10 away from my number, I'll tell you you're COLD")
print("If your guess FIRST is within 10 of my number, I'll tell you you're WARM")

print("If your next guesses is farther than your most recent guess, I'll say you're getting COLDER")
print("If your next guesses is closer than your most recent guess, I'll say you're getting WARMER")
print("LET'S PLAY!")
print('')



num = random.randint(1,100)

guesses = []

while True:

    g = int(input('give me num'))


    if not (1 <= g <= 100):
        print('OUT OF BOUND!')
        continue

    elif g == num:
        print(f'Congrats u did in {len(guesses) + 1} guesses')
        break

    guesses.append(g)


    if len(guesses) == 1:
        if abs(g-num) <= 10:
            print('warm')
        else:
            print('cold')

    else:
        if abs(g-num) <= abs(guesses[-2]-num):
            print('WARMER!')
        else:
            print('COLDER!')



```
