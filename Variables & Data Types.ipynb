### Dynamic Strings

If we want to concatenate a strings with dynamic values we can use the following notation:

```python
# we will add to a string another string and an integer. For the string we use %s, while for the integer we use %d
print("Hello, my name is %s. I'm %d years old" % ('David',26))
```

```python
# Why %s for an int also worked ?  --> because we can convert an int into a string
print("Hello, my name is %s. I'm %s years old" % ('David',26))
```

```python
#Error
# Why %d for a string didn't work ?  --> because we can't convert a string into an int
print("Hello, my name is %d. I'm %d years old" % ('David',26))
```

```python
# The values are implemented according to the order!!! (not by %s or %d)
print("Hello, my name is %s. I'm %s years old" % (26,'David'))
```

```python
# If you got a little bit confused, it's o.k
# We can alwas do the same thing as folows:
x = 'David'
y = 26

print("Hello, my name is " + x + " I'm " + str(y) +" years old")
```

### String Literals

<br>
<p style="font-size:23px"><code>'f'</code> string literal</p>
Format strings to include variables and expressions directly in the string

```python
# Using 'f' string literal
x = 'David'
y = 26

print(f"Hello, my name is {x} I'm {y} years old")
```

```python
# Using 'f' string literal
x = 'David'
y = 26

print(f"Hello, my name is {x + ' Cohen'} I'm {y + 4} years old")
```

<br></br>
<p style="font-size:23px"><code>'r'</code> string literal</p>
Treats backslashes as literal characters, ignoring escape sequences. Useful for regular expressions and file paths

```python
# Again! not what we wanted!!!

print("C:\path\to\file")
```

```python
# Works!

print(r"C:\path\to\file")
```

```python
# Again! not what we wanted!!!
print('\t')
```

```python
# Works!
print(r'\t')
```

### Strings Formating

<p style="font-size:20px">Used to format numbers, especially floating-point numbers, into a specific representation</p>

```python
f'the number pi = {3.1415926}'
```

<br>
<b><u>Round</u></b> Float to 2 Decimal Places

```python
# Round Float to 2 Decimal Places
# Syntax: f'{num:.2f}'  -  for rounding to 2 decimal places.

    # {} marks a replacement field
    # : introduces a format specifier
    # .2 specify the precision as 2, or any other number
    # f to format the number as a decimal number
    
f'the number pi = {3.1415926:.2f}'
```

```python
# Add Thousands Separator

f'Add Thousands Separator = {1234567.89:,}'
```

```python
# Round Float to 2 Decimal Places with Sign

f'Round Float to 2 Decimal Places with Sign = {3.1415926:+.2f}'
```

```python
# Add Thousands Separator
# Round Float to 2 Decimal
# Adds the $ sign

f'{1234567.896868:,.2f}'
```

```python
# Format as Percentage

f'Format as Percentage = {0.5146:.2%}'
```

## `Int` Methods

```python
a = 5
a
```

```python
type(a)
```

<code>.bit_length()</code> method returns the number of bits required to represent an integer in binary, excluding the sign and leading zeros.

```python
# The binary representation of 5 is 101
a.bit_length()
```

To convert the number 5 to binary:

- Divide 5 by 2, the quotient is 2, and the remainder is 1.
- Divide 2 by 2, the quotient is 1, and the remainder is 0.
- Divide 1 by 2, the quotient is 0, and the remainder is 1.

Reading the remainders from bottom to top gives us 101, which is the binary representation of the decimal number 5.

## NoneType

<br>
<p style="font-size:23px">Singleton Object</p>
<p style="font-size:20px"><code>None</code> is a singleton, meaning there is only one instance of it.
<br>This is why we use <code>is</code> to check for <code>None</code>.
<br>(e.g., <code>x is None</code>), not <code>==</code></p>

```python
x = None
```

```python
x
```

```python
type(x)
```

```python
x is None
```

```python
x == None
```

<p style="font-size:20px">both <code>x == None</code> and <code>x is None</code> will indeed evaluate to <code>True</code> when <code>x</code> is <code>None</code>.
<br>However, thereâs an important distinction in Python:</p>

- `x is None`: This checks if `x` and `None` refer to the exact same object (object identity). It's the recommended and more precise way to check for `None`.
- `x == None`: This checks if `x` is "equal" to `None` in terms of value (value equality), which might work here but is less precise, as some objects might define custom equality methods where `== None` could behave unexpectedly.

```python
x is not None
```
