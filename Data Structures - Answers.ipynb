### Exercise 24

 Given a dictionary where all values are unique, swap the keys and values.<br>
 For example, if the input dictionary is <code>{'a': 1, 'b': 2, 'c': 3}</code>, the output should be <code>{1: 'a', 2: 'b', 3: 'c'}</code>.

```python
d = {'a': 1, 'b': 2, 'c': 3}

# Swapping keys and values
swapped_d = dict(zip(d.values(), d.keys()))

swapped_d
```

## Python Sets

### Exercise 25

Consider the list: <code>A = [1, 2, 2, 1]</code>
<br>
and set <code>B = set([1, 2, 2, 1])</code>,
<br>
<br>
does <code>sum(A) = sum(B)</code>

```python
A = [1, 2, 2, 1]
B = set([1, 2, 2, 1])

print("the sum of A is:", sum(A))
print("the sum of B is:", sum(B))

sum(A) == sum(B)
```

### Exercise 26

### Exercise 27

Convert the list <code>[ "AC/DC", "Back in Black", "The Dark Side of the Moon"]</code> to a set name `album_set2`

```python
album_set2 = set([ "AC/DC", "Back in Black", "The Dark Side of the Moon"])
album_set2
```

### Exercise 28

Create a new set <code>album_set3</code> that is the union of <code>album_set1</code> and <code>album_set2</code>

```python
album_set3 =  album_set1 | album_set2

# another way

album_set3 = album_set1.union(album_set2)

album_set3
```

### Exercise 29

Find out if <code>album_set1</code> is a subset of <code>album_set3</code>:

```python
album_set1 - album_set3  # -- True - cause gives you an empty set
```

Another way


```python
album_set1.issubset(album_set3)
```
