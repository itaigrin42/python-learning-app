## Function can Return More than One Value

```python
def ranging(x):
    l = []
    
    while x >= 0:
        l.append(x)
        x -= 1
    
    return l
```

```python
ranging(5)
```

## Arbitrary Arguments, *args
Not sure how many arguments that will be passed into your function, add a * before the parameter<br>
The function will receive a tuple of arguments

```python
def the_biggest_num_one_arg(num):
    x=max(num)
    print(x)
```

```python
# Error
the_biggest_num_one_arg(5,6,7,8,9)
# The function expects only one arg
```

```python
# How can we use "the_biggest_num_one_arg" function with multiple arg ??
# We can put the args in a tupple
the_biggest_num_one_arg((5,6,7,8,9))
```

```python
def the_biggest_num(*num):
    x=max(num)
    print(x)
```

```python
the_biggest_num(5,6,7,8,9)
```

```python
def show_me_the_tupple(*num):
    return(num)

show_me_the_tupple(5,6,7,8,9,'a','b','c',("I am a ",'tupple'))
# "num" is a tupple
```

```python
t = show_me_the_tupple(5,6,7,8,9,'a','b','c')
type(t)
```

## keyword arguments ,**kwargs
A special syntax that allows you to pass a keyworded, variable-length argument dictionary to a function

```python
# Using Keywords
def the_biggest_num2(**num):
    x=num["big"]
    print('this is the biggest num ' + str(x))

the_biggest_num2(small=5,medium=6,big=7)

```

```python
# Using Keywords
def the_biggest_num2(**num):
    print(num)

the_biggest_num2(small=5,medium=6,big=7)
# "num" is a dictionary
```

```python
# Using Keywords
def greet(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

greet(name="Alice", greeting="Hello")
```

```python
# Using Keywords
def create_profile(name, age=30, **kwargs):
    profile = {'name': name, 'age': age}
    profile.update(kwargs)
    return profile


profile1 = create_profile("Alice")
profile2 = create_profile("Bob", age=40, location="New York", occupation="Engineer")

print(profile1)
print(profile2)
```

The <i><b><u>map()</u></b> </i> function executes a specified function for each item in an iterable. The item is sent to the function as a parameter.

## Lambda Functions

Lambda functions are intended as a shorthand for defining functions that can come in handy to write concise code without wasting multiple lines defining a function.
<br>
They are also known as anonymous functions, since they do not have a name unless assigned one.

```python
# Anonymous
# defining lambda function without assigning a name to the function
lambda x: x + 10

# left side --> parameter
# write side --> expression
```

```python
# Anonymous
# Activating lambda function on the int 30
(lambda x: x + 10)(30)
```

```python
# Assigning a name to the lambda function
y = lambda x: x + 10
```

```python
y
```

```python
y(30)
```

```python
# We can also do the same with "udf"
# But we have to give it a name
def increase10(x):
    return x + 10
```

```python
increase10(30)
```

### When Should You Use a Lambda Function?
You should use the lambda function to create simple expressions. For example, expressions that do not include complex structures such as if-else, for-loops, and so on.

So, for example, if you want to create a function with a for-loop, you should use a user-defined function.

### How to Use a Lambda Function with Iterables
An iterable is essentially anything that consists of a series of values, such as characters, numbers, and so on.

In Python, iterables include strings, lists, dictionaries, ranges, tuples, and so on. When working with iterables, you can use lambda functions in conjunction with two common functions: filter() and map().

```python
l = list(range(1,201))
l[0:3]
```

```python
# Anonymous
filter(lambda x: x % 2 == 0, l)
```

```python
l_zugi = list(filter(lambda x: x % 2 == 0, l))
l_zugi[0:3]
```

```python
# Anonymous
l_power = list(map(lambda x: x**2, l))
l_power[0:3]
```

```python
def bmi(weight, height):
    
    if (not isinstance(height, float)):
        print('''Did you mean %f ? Please enter the height in meters and not in centimeters.''' % (float(height)/100))
    
    else:
        res = weight / height**2
    
        return(res)
```

```python
def bmi(weight, height):
    
    
    try:
        if(not  isinstance(height, float)):
            print('''Did you mean %f ? Please enter the height in meters and not in centimeters.''' % (float(height)/100))
        else:
            res = weight / height**2
            return(res)
    
    
    
    
    except ZeroDivisionError:
        print("Please give a height different of zero")
    except TypeError:
        print("Please enter only numeric value")
    except:
        print("An unexpected error ocurred!")
```
